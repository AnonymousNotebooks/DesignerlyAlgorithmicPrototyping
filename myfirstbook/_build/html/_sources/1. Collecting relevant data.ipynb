Before collecting this data, we needed to identify the channels through which an ML system could realistically harvest data within the identified domestic practice. These channels could be, for instance: (1) device sensors that track user interactions; (2) wearable sensors that monitor biometric data, the user’s current activity, or their location; (3) fixed sensors that detect environmental data, such as temperature, occupancy, or light intensity; (4) services that collect information about their users' behavior, such as purchase patterns or resource consumption patterns. The data that was collected for the grocery automation case study was in the form of grocery receipts(figure~\ref{fig:receipts}).
